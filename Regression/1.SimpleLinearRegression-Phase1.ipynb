```python
import pandas as pd #pandas is a global file for to access the dataset csv files, pandas functions reassign to pd
```

```python
dataset = pd.read_csv("Salary_Data.csv"); #Read the data from salary_data.csv file using read.csv function as imported from pandas as pd and assign to dataset variable
```

```python
dataset #View the data using dataset variable
```

```python
independent = dataset[["YearsExperience"]] #Assign the input column data to particicular variable
```

```python
dependent = dataset[["Salary"]] #Assihn the output column data to particicular individual variable
```

```python
from sklearn.model_selection import train_test_split #Imprt the train_test_split function from model_selection class from sklearn package
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0) #Split the train and test data from independent and dependent variable.
#test_size is a rasio for split the data
#random_state=0 for default
#train_test_split function return the list of items, list total count is 4
#first item is input train data
#sceond item is input test data
#third item is output train data
#fourth item is output test data
```

from sklearn.linear_model import LinearRegression

```python
from sklearn.linear_model import LinearRegression #Imported the LinearRegression class from linear_model class from sklearn package
```

```python
regressor = LinearRegression() #Call the LinearRegression class & assign to regressor variable
#regressor is a Object
```

```python
regressor.fit(x_train, y_train) #When we call fit method with x_train and y_train, that time Learning phase is process and Created the model
```

```python
weight = regressor.coef_ #Read & assign the weight value form regressor.coef_
```

```python
weight #View the weight value
```

```python
bais = regressor.intercept_ #Read & assign the bais value from regressor.intercept_
```

```python
bais #View the bais value
```

```python
y_pred = regressor.predict(x_test) #Predict the values from x_test and store y_pred variable
```

```python
y_pred #View the y_pred values
```

```python
from sklearn.metrics import r2_score #Import r2_score method from metrics class from sklearn package
```

```python
r_score = r2_score(y_test, y_pred) #Calulate the r_score value using r2_score and pass y_test, y_test data
```

```python
r_score #View r_score value
```

```python
import pickle #Import the pickle package for Model saving
```

```python
filename = "finalized_model_linear.sav" #Create the dummy name for Model save
```

```python
pickle.dump(regressor, open(filename,'wb')) #pickle.dump method process to save the regressor model to assigned filename with write binary
```

```python
loaded_model = pickle.load(open(filename,'rb')) #Saved regressor model is loaded in loaded_model variable with read binary 
```

```python
result = loaded_model.predict([[15]]) #Predict the 15 years experence salary and save to result variable
```
